# State EPD

```python
def pairwise_dist(x):
    dist_matrix = cdist(x, x)
    return dist_matrix

def epd(pos1, pos2):
    pos1_pd = pairwise_dist(pos1)
    pos2_pd = pairwise_dist(pos2)
    
    pd = np.square(pos1_pd - pos2_pd).mean()
    mean_pd, std_pd = pd.mean().item(), pd.std().item()
    return mean_pd * 1000, std_pd

def epd_two_states(state1, state2):
    state1_file = f'../data/{molecule}/{state1}.pdb'
    state2_file = f'../data/{molecule}/{state2}.pdb'
    file_state1 = md.load(state1_file)
    file_state2 = md.load(state2_file)
    
    return epd(file_state1.xyz[0], file_state2.xyz[0])
```

```python
state1 = "c5"
state2 = "c7ax"

epd_two_states(state2, state1)
```

```python
print(epd_two_states("c5", "c7ax")
print(epd_two_states("c5", "al")
print(epd_two_states("c7ax", "c5")
print(epd_two_states("c7ax", "ar")
```
